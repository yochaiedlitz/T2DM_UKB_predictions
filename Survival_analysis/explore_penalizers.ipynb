```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
!ls /home/edlitzy/UKBB_Tree_Runs/For_article/revision/SA/Five_BT_SA/training_Results/
training_dir="/home/edlitzy/UKBB_Tree_Runs/For_article/revision/SA/SA_Five_BT/training_Results/"

def penalizer_impact(training_dir):
    csv_list=[]
    rel_files=[x for x in os.listdir(training_dir) if x.startswith("AUC_APS_results") and x.endswith(".csv")]
    for x in rel_files:
        tmp_path=os.path.join(training_dir,x)
        tmp_csv=pd.read_csv(tmp_path, index_col=0)
        csv_list.append(tmp_csv)
    results_csv=pd.concat(csv_list)
    fig,axs=plt.subplots(4,1,figsize=(9,24))
    results_csv["l1_ratio"].hist(bins=100,ax=axs[0])
    axs[0].set_title("l1_ratio hist")
    results_csv.plot(x="penalizer",y="AUC",ax=axs[1])
    axs[1].set_title("Penalizer impact")
    results_csv["l1_panelizer"]=results_csv["l1_ratio"]*results_csv["penalizer"]
    results_csv.plot(x="l1_panelizer",y="AUC",ax=axs[2])
    axs[2].set_title("l1_panelizer impact")
    results_csv["l2_panelizer"]=(1-results_csv["l1_ratio"])*results_csv["penalizer"]
    results_csv.plot(x="l2_panelizer",y="AUC",ax=axs[3])
    axs[3].set_title("l2_panelizer impact")

    plt.show()
    return results_csv
```

```python
training_dir="/home/edlitzy/UKBB_Tree_Runs/For_article/revision/SA/SA_Five_BT/training_Results/"
penalizer_impact(training_dir)
```
